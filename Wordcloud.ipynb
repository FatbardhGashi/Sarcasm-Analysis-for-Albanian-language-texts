```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud, ImageColorGenerator

from os import path
from PIL import Image
```

```python
data = pd.read_csv("../Dataseti/trainset.csv")
```

```python
text = data['Comment']
komentet = []
for i in range(len(text)):
    komentet.append(str(text[i]))

stringKomentet = ""

for i in range(len(komentet)):
    stringKomentet = " ".join(komentet)
```

```python
wordcloud = WordCloud(width=800, height=500, max_words=100, background_color="white", min_word_length=5, stopwords=["eshte", "duhet", "https"]).generate(stringKomentet)

plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()

wordcloud.to_file("wordcloud.jpeg")
```

```python
# create a dictionary of word frequencies
text_dictionary = wordcloud.process_text(stringKomentet)

# sort the dictionary
word_freq={k: v for k, v in sorted(text_dictionary.items(),reverse=True, key=lambda item: item[1])}

#use words_ to print relative word frequencies
rel_freq=wordcloud.words_

#print results
print(list(word_freq.items())[:10])
```
