```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../Dataseti/trainset.csv")
```

```python
#Klasifikimi 1
print("Anotuesi 1: ")
print("Klasifikimet sarkazëm: ", (data['Annotator1'] == 1).sum())
print("Klasifikimet jo sarkazëm: ", (data['Annotator1'] == 0).sum())
```

```python
data['Annotator1'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(7,7), explode = (0, 0.02),
                                                     labels=['Jo sarkazëm', 'Sarkazëm'],
                                                     colors = ("#E74C3C", "#2E86C1"))
plt.title('Anotuesi 1')
```

```python
#Klasifikimi 2
print("Anotuesi 2: ")
print("Klasifikimet sarkazëm: ", (data['Annotator2'] == 1).sum())
print("Klasifikimet jo sarkazëm: ", (data['Annotator2'] == 0).sum())
```

```python
data['Annotator2'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(7,7), explode = (0, 0.02),
                                                     labels=['Jo sarkazëm', 'sarkazëm'],
                                                    colors = ("#E74C3C", "#2E86C1"))#,  "grey", "indigo", "beige") )
plt.title('Anotuesi 2')
```

```python
#Klasifikimi 3
print("Anotuesi 3: ")
print("Klasifikimet sarkazëm: ", (data['Annotator3'] == 1).sum())
print("Klasifikimet jo sarkazëm: ", (data['Annotator3'] == 0).sum())
```

```python
data['Annotator3'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(7,7), explode = (0, 0.02),
                                                     labels=['Jo sarkazëm', 'sarkazëm'],
                                                    colors = ("#E74C3C", "#2E86C1"))#,  "grey", "indigo", "beige") )
plt.title('Anotuesi 3')
```

```python
#Klasifikimi final
print("Klasifikimi final: ")
print("Klasifikimet sarkazëm: ", (data['Final annotation'] == 1).sum())
print("Klasifikimet jo sarkazëm: ", (data['Final annotation'] == 0).sum())
```
